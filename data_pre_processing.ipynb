```python
new_keys = ['Wikipedia movie ID','Movie name','Movie release date','Movie box office revenue','Movie runtime','Movie languages','Movie countries','Movie genres']
columns = ['Wikipedia movie ID','Movie name','Movie release date','Movie box office revenue','Movie runtime']
new_movie_metadata = pandas.DataFrame(index=range(len(movie_metadata)), columns=new_keys)
for column in columns:
    new_movie_metadata.loc[:,column] = movie_metadata.loc[:,column]
```

```python
temp1 = []
temp2 = []
temp3 = []
for i in range(len(movie_metadata)):
    temp1.append(list(ast.literal_eval(movie_metadata.loc[i,'Movie languages']).values()))
    
    temp2.append(list(ast.literal_eval(movie_metadata.loc[i,'Movie countries']).values()))
    
    temp3.append(list(ast.literal_eval(movie_metadata.loc[i,'Movie genres']).values()))
```

```python

new_movie_metadata.at[:,'Movie languages'] = temp1[:]   
new_movie_metadata.at[:,'Movie countries'] = temp2[:]
new_movie_metadata.at[:,'Movie genres'] = temp3[:]
```

```python
list_of_genre = []
for i in range(len(temp3)):
    list_of_genre.extend(temp3[i][:])    
list_of_genre = list(set(list_of_genre))    
```

```python
#get a dictionary with each genre and its appearence times 
list_sequence=[]
for i in range(len(movie_metadata)):
    list_sequence.extend(list(ast.literal_eval(movie_metadata.loc[i,'Movie genres']).values()))
genre_count={}
for item in list_of_genre:
    genre_count[item]=list_sequence.count(item)
genre_count
```

```python
#get a list of each genre and its apprearence times, order in descending
genre_count_list=[]
for item in list_of_genre:
    genre_count_list.append([item,list_sequence.count(item)])
import operator
sorted(genre_count_list,key=operator.itemgetter(1),reverse=True)
```

We should choose labels which are most frequencent and typical. Labels should also be distinct enough from each other. According to such principle we sort the genres in the followings:'Drama'(Drama,Romantic drama,Comedy-drama,Family Drama),'Comedy'(Comedy,Comedy film,Romantic comedy,Comedy-drama),'Romance'(Romance Film, Romantic drama,Romantic comedy),'Action'(Action,Action/Adventure,Action Thrillers),‘Horror’(Horror), 'Thriller'(Thriller,Crime Thriller,Action Thrillers),'Crime'(Crime Fiction,Crime Thriller),'Documentary'(Documentary),'Adventure'(Adventure,Action/Adventure),'Family'(Family Film,Family Drama),'Music'(Musical,Music),'Animation'(Animation),'Mystery'(Mystery),'Science Fiction'(Science Fiction),'Fantasy'(Fantasy),'War'(War film),'Western'(Western),'Sports'(Sports)

```python
Genre_classifier={'Drama':['Drama'],'Comedy':['Comedy'],'Romance Film':['Romance'],'Action':['Action'],'Thriller':['Thriller'],'Crime Fiction':['Crime'],'Documentary':['Documentary'],'Horror':['Horror'],'Adventure':['Adventure'],'Family Film':['Family'],'Action/Adventure':['Action','Adventure'],'Comedy film':['Comedy'],'Musical':['Music'],'Animation':['Animation'],'Romantic drama':['Romance','Drama'],'Mystery':['Mystery'],'Science Fiction':['Science Fiction'],'Fantasy':['Fantasy'],'Romantic comedy':['Romance','Comedy'],'War film':['War'],'Western':['Western'],'Crime Thriller':['Crime','Thriller'],'Comedy-drama':['Comedy','Drama'],'Sports':['Sports'],'Music':['Music']}
```

```python
#define new values for Movie genre

temp3_1=[]
for labels in temp3:
    l=labels[:]
    for label in labels:
        if label in Genre_classifier:
            l.remove(label)
            l.extend(item for item in Genre_classifier[label])
        else:
            l.remove(label)
    l=list(set(l))
    temp3_1.append(l)
print len(temp3_1)

new_movie_metadata.at[:,'Movie genres']=temp3_1[:]
print new_movie_metadata.loc[:,'Movie genres']
```
